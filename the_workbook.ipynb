### Installing & Importing the libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress

print("importing the required libraries complete")
```

### Importing the data

```python
df = pd.read_csv('./epa-sea-level.csv')
df.head()
```

### Analysing

```python
plt.scatter(
    df['Year'],
    df['CSIRO Adjusted Sea Level']
)
plt.title("Change in sealevel according to the years")
plt.xlabel('Year')
plt.ylabel('CSIRO Adjusted Sea level')
plt.show
```

```python
#extracting data 
years= df['Year']
sea_levels= df["CSIRO Adjusted Sea Level"]
#performing linear regression
slope, intercept, r_value, p_value, std_err = linregress ( years, sea_levels)

#predicting the future sea level
future_years = np.arange(min(years),2051)
predicted_sea_levels = slope * future_years + intercept
plt.plot(future_years, predicted_sea_levels, label = "Line of Best Fit", color="red")
plt.xlabel("Years")
plt.ylabel("Sea level (inches)")
plt.title("Rise in Sea Level")
```
